```python
from numba import njit

import random

@njit
def monte_carlo_pi(nsamples):
    acc = 0
    for i in range(nsamples):
        x = random.random()
        y = random.random()
        if (x ** 2 + y ** 2) < 1.0:
            acc += 1
    return 4.0 * acc / nsamples
```

```python
monte_carlo_pi(1000)
```

```python
monte_carlo_pi.inspect_disasm_cfg(monte_carlo_pi.signatures[0])
```

```python
@njit
def foo(x):
    return x + 1 if x < 3 else 3*x
```
